```python
import numpy as np
import pylab as plt
import seaborn as sns
```

```python
no_headcase_AP = np.loadtxt('sub-01_ses-01_task-Stroop_acq-cf0AP_run-03_bold_fd.txt', skiprows=1)
no_headcase_PA = np.loadtxt('sub-01_ses-02_task-Stroop_acq-cf0PA_run-06_bold_fd.txt', skiprows=1)
no_headcase = np.concatenate([no_headcase_AP, no_headcase_PA])

tight_cf_AP = np.loadtxt('sub-01_ses-01_task-Stroop_acq-cf1AP_run-01_bold_fd.txt', skiprows=1)
tight_cf_PA = np.loadtxt('sub-01_ses-01_task-Stroop_acq-cf1PA_run-02_bold_fd.txt', skiprows=1)
tight_cf = np.concatenate([tight_cf_AP, tight_cf_PA])

normal_cf_AP = np.loadtxt('sub-01_ses-02_task-Stroop_acq-cf2AP_run-01_bold_fd.txt', skiprows=1)
normal_cf_PA = np.loadtxt('sub-01_ses-02_task-Stroop_acq-cf2PA_run-02_bold_fd.txt', skiprows=1)
normal_cf = np.concatenate([normal_cf_AP, normal_cf_PA])
```

```python
sns.distplot(no_headcase, hist=False, label='normal padding')
sns.distplot(tight_cf, hist=False, label='tight caseforge')
sns.distplot(normal_cf, hist=False, label='normal caseforge')
plt.xlabel('FramewiseDisplacement [mm]')
```
